```haskell
b = parseProgram $ unlines [
  "@1            ",
  "   MOV 0 ACC  ",
  "   SUB IN     ",
  "&: ADD @2     ",
  "   MOV -2 @2  ",
  "   JLZ &      ",
  "   MOV @2 @2  ",
  "   MOV ACC @2 ",
  "              ",
  "@2            ",
  "&: MOV ACC @1 ",
  "   SUB @1     ",
  "   JGZ &      ",
  "   MOV @1 OUT "]

run b [36] def (renderProgram . (!! 25))

run b (map (^2) [1,4]) def log
```
